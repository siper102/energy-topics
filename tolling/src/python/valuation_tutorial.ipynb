# Tolling Agreement Valuation

This notebook demonstrates the valuation of a **Gas-Fired Tolling Agreement** using a standardized Python API backed by a high-performance **Rust** engine.

## 1. Introduction

A tolling agreement is a contract where the buyer (the "Tolling Party") supplies fuel (Natural Gas) to a power plant and receives electricity (Power) in return, paying for the conversion capacity. The plant owner is responsible for the technical operation.

The value of this agreement depends on the **Spark Spread** (the difference between electricity price and the cost of gas required to generate it).

### The Model
We model the asset prices under the risk-neutral measure $\mathbb{Q}$:

*   **Natural Gas ($G_t$):** Geometric Brownian Motion (GBM).
*   **Electricity ($P_t$):** Mean-Reverting Jump Diffusion (MRJD).
*   **Correlation:** The drivers of Gas and Power are correlated with coefficient $\rho$.

### Architecture
We use the **`ValuationModel`** interface (`src/python/interfaces`) which abstracts the underlying calculation engine. In this case, we use `TollingModel`, which delegates the heavy lifting to our optimized Rust library.

```python
import sys
import os
import time
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure we can import from src/python
notebook_dir = Path(os.getcwd())
src_python_dir = notebook_dir.parent
if str(src_python_dir) not in sys.path:
    sys.path.append(str(src_python_dir))

# Import the Model Wrapper
from python.models.tolling import TollingModel
import tolling_agreement_valuation

# Set plotting style
sns.set_theme(style="whitegrid")
```

## 2. Data Setup

We load the input parameters and forward curves into Python objects (Dictionaries and DataFrames). This demonstrates the flexibility of the API—it works with in-memory data, not just files.

```python
model_params = tolling_agreement_valuation.ModelParameters(
    sigma_g = 0.4,
    sigma_p =  0.5,
    kappa = 50.0,
    lambda_j = 5.0,
    mu_j = 0.5,
    sigma_j = 0.3,
    rho = 0.6
)
unit_params = [
    tolling_agreement_valuation.UnitParameter(
        heat_rate = 1.67,
        capacity = 400,
        start_up_costs = 15000
    ),
    tolling_agreement_valuation.UnitParameter(
        heat_rate = 3.33,
        capacity = 100,
        start_up_costs = 2000
    ),
]

print("Model Parameters (Rust Object):")
print(f"Sigma G: {model_params.sigma_g}, Sigma P: {model_params.sigma_p}...")
```

```python
dates = pd.date_range(
    start="2025-01-01 00:00:00",
    end="2025-12-31 23:00:00",
    freq="h"
)

hours = np.arange(len(dates))
t_years = hours / (365.0 * 24.0)
```

```python
base_price_gas = 3.50
seasonality_gas = 0.50

curve = base_price_gas + seasonality_gas * np.sin(2 * np.pi * t_years)
gas_curve =  pd.DataFrame(
        {
            "date": dates,
            "price": curve,
        }
).set_index("date")
```

```python
base_price_power = 45
seasonality_power = 10

curve = base_price_power + seasonality_power * np.sin(2 * np.pi * t_years)
power_curve = pd.DataFrame(
    {
        "date": dates,
        "price": curve,
    }
).set_index("date")
```

```python
fig, ax = plt.subplots(figsize=(12, 5))
gas_curve['price'].plot(ax=ax, label='Gas Forward ($/MMBtu)')
(power_curve['price'] / 10).plot(ax=ax, label='Power Forward ($/MWh / 10)', secondary_y=True)

ax.set_title("Hourly Forward Curves (2025)")
ax.set_ylabel("Gas Price")
ax.right_ax.set_ylabel("Power Price (Scaled)")
ax.legend(loc='upper left')
plt.show()
```

## 3. Simulation (via Standard API)

We instantiate the `TollingModel`, load our data, and request sample paths. The model handles the communication with the Rust backend transparently.

```python
# Initialize Model
model = TollingModel()

# Load Data
model.load_parameters(model_params, unit_params)
model.load_forward_curves({
    'gas': gas_curve, 
    'power': power_curve
})

NUM_PATHS = 10_000

start_time = time.time()

# Generate Paths
sim_paths = model.get_sample_paths(NUM_PATHS)

end_time = time.time()
print(f"Simulation of {NUM_PATHS} paths took {end_time - start_time:.4f} seconds.")
print(f"Output Shape: {sim_paths.shape} (Paths, Assets, Hours)")
```

## 4. Visualization

We visualize the simulated paths to ensure they behave as expected.

```python
# Extract Gas (index 0) and Power (index 1)
gas_paths = sim_paths[:, 0, :]
power_paths = sim_paths[:, 1, :]

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10), sharex=True)

# --- Gas Plot ---
ax1.plot(gas_paths.T[:, :50], color='teal', alpha=0.1, linewidth=0.5)
ax1.plot(gas_paths.mean(axis=0), color='black', linewidth=2, linestyle='--', label='Simulated Mean')
ax1.plot(gas_curve['price'].values, color='red', linewidth=1, linestyle=':', label='Forward Curve')

ax1.set_title("Simulated Gas Prices (GBM)")
ax1.set_ylabel("Price ($/MMBtu)")
ax1.legend()

# --- Power Plot ---
ax2.plot(power_paths.T[:, :50], color='orange', alpha=0.1, linewidth=0.5)
ax2.plot(power_paths.mean(axis=0), color='black', linewidth=2, linestyle='--', label='Simulated Mean')
ax2.plot(power_curve['price'].values, color='blue', linewidth=1, linestyle=':', label='Forward Curve')

ax2.set_title("Simulated Power Prices (MRJD)")
ax2.set_ylabel("Price ($/MWh)")
ax2.set_xlabel("Hour of Year")
ax2.legend()

plt.tight_layout()
plt.show()
```

## 5. Daily Profit Calculation

We calculate the non-discounted daily profits for each simulation path.

```python
VALUATION_PATHS = 10_000
R_RATE = 0.02

print(f"Calculating profits with {VALUATION_PATHS} paths...")

# 1. Calculate undiscounted profits (r=0)
daily_profits_raw = model.calculate_daily_profits(VALUATION_PATHS, risk_free_rate=0.0)

# 2. Calculate discounted profits (r=0.02)
daily_profits_discounted = model.calculate_daily_profits(VALUATION_PATHS, risk_free_rate=R_RATE)

print(f"------------------------------------------------")
print(f"Mean daily profit (undiscounted): €{daily_profits_raw.mean():,.2f}")
print(f"Mean daily profit (discounted):   €{daily_profits_discounted.mean():,.2f}")
print(f"------------------------------------------------")
```

## 5.1. Discounting and Present Value

Now we apply the risk-free discount rate to calculate the present value of the tolling agreement. Daily profits are discounted back to present value using the risk-free rate $r$ from the model parameters.

The discount factor for each day is calculated as:

$$\text{DF}(t) = e^{-r \cdot t}$$

where $t = \frac{\text{day} + 1}{365}$ (assuming cash flows occur at the end of each day).

```python
# The total Present Value (PV) is the sum of discounted daily profits.
pv_per_path = daily_profits_discounted.sum(axis=1)
raw_sum_per_path = daily_profits_raw.sum(axis=1)

expected_pv = pv_per_path.mean()
pv_std = pv_per_path.std()
pv_confidence_interval = 1.96 * pv_std / np.sqrt(len(pv_per_path))

print(f"=== Present Value Analysis ===")
print(f"Discount rate: {R_RATE * 100:.2f}%")
print(f"------------------------------------------------")
print(f"Expected PV (Mean): €{expected_pv:,.2f}")
print(f"PV Std Dev: €{pv_std:,.2f}")
print(f"95% CI: €{expected_pv:,.2f} ± €{pv_confidence_interval:,.2f}")
print(f"------------------------------------------------")
print(f"\nComparison:")
print(f"  Undiscounted total profit (mean): €{raw_sum_per_path.mean():,.2f}")
print(f"  Discounted PV (mean):             €{expected_pv:,.2f}")
print(f"  Discount impact: {100 * (1 - expected_pv / raw_sum_per_path.mean()):.2f}%")
```

```python
# Visualize the distribution of present values
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Histogram of PV distribution
axes[0].hist(pv_per_path, bins=50, color='steelblue', alpha=0.7, edgecolor='black')
axes[0].axvline(expected_pv, color='red', linestyle='--', linewidth=2, label=f'Mean: €{expected_pv:,.0f}')
axes[0].axvline(expected_pv - pv_confidence_interval, color='orange', linestyle=':', linewidth=2, label='95% CI')
axes[0].axvline(expected_pv + pv_confidence_interval, color='orange', linestyle=':', linewidth=2)
axes[0].set_title('Distribution of Present Values Across Paths')
axes[0].set_xlabel('Present Value (€)')
axes[0].set_ylabel('Frequency')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Impact of discounting over time
cumulative_undiscounted = daily_profits_raw.mean(axis=0).cumsum()
cumulative_discounted = daily_profits_discounted.mean(axis=0).cumsum()
num_days = len(cumulative_undiscounted)

axes[1].plot(cumulative_undiscounted, label='Undiscounted', linewidth=2, color='teal')
axes[1].plot(cumulative_discounted, label='Discounted (PV)', linewidth=2, color='navy')
axes[1].fill_between(range(num_days), cumulative_undiscounted, cumulative_discounted, 
                      alpha=0.3, color='orange', label='Discount Impact')
axes[1].set_title('Cumulative Value Over Time')
axes[1].set_xlabel('Day of Year')
axes[1].set_ylabel('Cumulative Value (€)')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 6. Profit Distribution Analysis

Let's analyze the distribution of daily profits across all paths and days.

```python
# Aggregate statistics
mean_daily_profit_per_path = daily_profits_discounted.mean(axis=1)  # Average across days for each path
mean_daily_profit_per_day = daily_profits_discounted.mean(axis=0)   # Average across paths for each day

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. Distribution of mean daily profits per path
axes[0, 0].hist(mean_daily_profit_per_path, bins=50, color='teal', alpha=0.7, edgecolor='black')
axes[0, 0].axvline(mean_daily_profit_per_path.mean(), color='red', linestyle='--', linewidth=2, label='Overall Mean')
axes[0, 0].set_title('Distribution of Path-Average Daily Profits')
axes[0, 0].set_xlabel('Average Daily Profit (€)')
axes[0, 0].set_ylabel('Frequency')
axes[0, 0].legend()

# 2. Time series of expected daily profit
axes[0, 1].plot(mean_daily_profit_per_day, color='navy', linewidth=1)
axes[0, 1].set_title('Expected Daily Profit Over Time')
axes[0, 1].set_xlabel('Day of Year')
axes[0, 1].set_ylabel('Expected Daily Profit (€)')
axes[0, 1].grid(True, alpha=0.3)

# 3. Heatmap of sample paths (first 100 paths, every 7th day)
sample_paths = daily_profits_discounted[:100, ::7]
im = axes[1, 0].imshow(sample_paths, aspect='auto', cmap='RdYlGn', interpolation='nearest')
axes[1, 0].set_title('Daily Profits Heatmap (Sample)')
axes[1, 0].set_xlabel('Day (every 7th)')
axes[1, 0].set_ylabel('Path')
plt.colorbar(im, ax=axes[1, 0], label='Profit (€)')

# 4. Box plot of daily profits by month
days_per_month = 365 // 12
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
monthly_data = []
for i in range(12):
    start_day = i * days_per_month
    end_day = (i + 1) * days_per_month if i < 11 else 365
    monthly_data.append(daily_profits_discounted[:, start_day:end_day].flatten())

axes[1, 1].boxplot(monthly_data, labels=months, showfliers=False)
axes[1, 1].set_title('Daily Profit Distribution by Month')
axes[1, 1].set_xlabel('Month')
axes[1, 1].set_ylabel('Daily Profit (€)')
axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# Summary statistics
print(f"\n=== Summary Statistics ===")
print(f"Total paths × days: {daily_profits_discounted.size:,}")
print(f"Positive profit days: {(daily_profits_discounted > 0).sum():,} ({100 * (daily_profits_discounted > 0).sum() / daily_profits_discounted.size:.1f}%)")
print(f"Min daily profit: €{daily_profits_discounted.min():,.2f}")
print(f"Max daily profit: €{daily_profits_discounted.max():,.2f}")
```

## 7. Greeks Analysis (AAD)

We calculate the sensitivities of the Tolling Agreement Value to input parameters using Adjoint Algorithmic Differentiation (AAD).

*   **Delta ($\Delta$):** Sensitivity to the initial forward curve prices ($dP/dF_t$).
*   **Vega ($\nu$):** Sensitivity to volatilities ($dP/d\sigma$).

AAD allows us to compute these sensitivities efficiently in a single backward pass.

```python
GREEK_CALCULATION_PATHS = 10000

print(f"Calculating Greeks with {GREEK_CALCULATION_PATHS} paths...")
start_time = time.time()

greeks = model.calculate_greeks(GREEK_CALCULATION_PATHS , risk_free_rate=R_RATE)

end_time = time.time()
print(f"Calculation took {end_time - start_time:.4f} seconds.")

print(f"------------------------------------------------")
print(f"Vega Gas:   €{greeks.vega_gas:,.2f} / %")
print(f"Vega Power: €{greeks.vega_power:,.2f} / %")
print(f"------------------------------------------------")

# Visualize Deltas
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(dates, greeks.delta_gas, label=r'Delta Gas ($dPV/dG_t$)', color='red')
ax.plot(dates, greeks.delta_power, label=r'Delta Power ($dPV/dP_t$)', color='blue')

ax.set_title("Delta Profile (Sensitivity to Forward Curve Shocks)")
ax.set_ylabel("Sensitivity (€ per unit change)")
ax.set_xlabel("Date")
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
